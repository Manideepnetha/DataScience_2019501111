```python
import pandas as pd
import numpy as np
import random
import glob
import csv, os
```

```python
def is_file_empty(file_path):
    return os.path.exists(file_path) and os.stat(file_path).st_size == 0
```

```python
count = 0
df = pd.DataFrame()
for files in glob.glob('F:\\Data Science\\DataScience_2019501111\\DADV\\Assignment 2\\*.csv'):
    if count == 30:
        break
    f=open(files, 'r')
    name = os.path.basename(f.name).replace(".csv","")
    if is_file_empty("F:\\Data Science\\DataScience_2019501111\\DADV\\Assignment 2\\"+name+".csv"):
        continue
    data = pd.read_csv(f)
    gain = (data.at[data.shape[0]-1, "Close Price"] / data.at[0,"Close Price"]) - 1
#     gain = (close[len(close)-1]/close[0])-1
    row = {'Code': name, "Gain": gain}
    df = df.append(row, ignore_index=True)
    count = count + 1
    f.close()
print("done")
```
